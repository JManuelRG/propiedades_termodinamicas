## IMPORTANCIÓN Y CREACIÓN DE MÓDULOS

Un módulo es un **archivo que contiene definiciones de funciones, claves y variables** que se pueden utilizar en otros programas.
 Su importación nos permite acceder a la funcionalidad definida en otros archivos y **reutilizar el código** de manera eficiente. Así mismo, podemos crear nuestros propios módulos para **organizar y modulizar** nuestro código.

 ## Importación de módulos

 Se utiliza la declaración **import**. Podemos importar un módulo completo o funciones específicas.



```python
import math

resultado = math.sqrt(25)
print(resultado)
print("Se importó módulo math, con declaración sqrt para calculo de la raíz cuadrada de 25")
```

**FUNCIONES ESPPECÍFICAS**

Utilizando **sintaxis** from módulo **import** función.

```python
from math import sqrt

resultado = sqrt(25)
print(resultado)
```
